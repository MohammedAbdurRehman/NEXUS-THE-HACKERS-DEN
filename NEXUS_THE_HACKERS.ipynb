## **Workshop on Linux CLI Basics: User, Kernel, and Command Mastery**

### **Theoretical Overview**
#### **1. User-Kernel Infrastructure**
- **Objective**: Explain the basic architecture of how user input is processed in Linux.
  
  - **User Space**: Where applications like the terminal (CLI) or graphical user interfaces (GUI) run.
  - **Kernel Space**: The core part of the OS that interacts directly with the hardware.

  **Flow Example**:
  - User enters a command in CLI or interacts via GUI.
  - The **shell** receives and interprets the command.
  - The shell passes the command to the **kernel**, which performs the necessary action (e.g., accessing the file system, executing a program).

  **Key Concept**: User → Shell → Kernel → Hardware

#### **2. Kernel vs Shell**
- **Kernel**: The core part of the OS that manages hardware and system resources. It handles requests from the shell.
  
- **Shell**: A program that interprets and executes user commands. There are different shells available in Linux, such as **Bash**, **Zsh**, etc.

  **Example**: When a user enters the command `ls`, the shell interprets it, and the kernel retrieves the directory contents.


### **Differences Between Windows and Linux OS**
**1. Architecture & Design**

**Windows:**

- Proprietary software with a graphical interface as the primary user interface.
- It has a monolithic kernel but includes user-friendly features for desktop use.
- File system: NTFS (New Technology File System).

**Linux:**

- Open-source OS, giving users full control over the system.
- It uses a monolithic kernel but is highly customizable for different environments (servers, desktops, etc.).
- File systems: EXT4, XFS, and others.

**2. User Interface**

- **Windows:** Primarily GUI-based, designed for ease of use for end users. Windows users interact mainly through windows, icons, and menus.

- **Linux:** Has both GUI (via desktop environments like GNOME, KDE) and CLI. CLI is a core component in Linux, giving users more control and flexibility.

**3. Command Line Usage**

- **Windows:** Uses Command Prompt or PowerShell, but they are less versatile compared to Linux's shell environments.

- **Linux:** Bash (Bourne Again Shell) and other shells are powerful and widely used for system administration, scripting, and automation.

**4. Security**

- **Windows:** Frequently targeted by malware due to its popularity and more closed nature. Security relies on third-party tools like antivirus programs.

- **Linux:** Generally considered more secure due to its open-source nature. Security patches are more frequent, and Linux distributions offer built-in security features.

### **Basic Linux Commands**
This section introduces key Linux commands, their structure, and usage.

#### **1. File Operations**
- **Create a File**:
  - `touch <filename>`: Creates an empty file.
    - Example: `touch file1.txt`
  
- **Copy a File**:
  - `cp <source> <destination>`: Copies a file from the source path to the destination path.
    - Example: `cp file1.txt /home/user/Documents/`
  
- **Move/Rename a File**:
  - `mv <source> <destination>`: Moves or renames a file.
    - Example (move): `mv file1.txt /home/user/Documents/`
    - Example (rename): `mv file1.txt file2.txt`
  
- **Remove a File**:
  - `rm <filename>`: Removes a file.
    - Example: `rm file1.txt`
  
#### **2. Directory Operations**
- **Create a Directory**:
  - `mkdir <directory_name>`: Creates a new directory.
    - Example: `mkdir my_folder`
  
- **Remove a Directory**:
  - `rmdir <directory_name>`: Removes an empty directory.
    - Example: `rmdir my_folder`
  
- **Remove a Directory with Files**:
  - `rm -r <directory_name>`: Recursively removes a directory and all its contents.
    - Example: `rm -r my_folder`
  
#### **3. Navigating Directories**
- **Current Directory**:
  - `pwd`: Prints the current working directory.
    - Example: `/home/user/Documents`
  
- **List Files**:
  - `ls`: Lists files in the current directory.
    - Example: `ls`
  
- **Change Directory**:
  - `cd <directory_path>`: Changes the current directory.
    - Example: `cd /home/user/Documents`
  
#### **4. Absolute vs Relative Paths**
- **Absolute Path**: A complete path from the root `/` directory.
  - Example: `/home/user/Documents/file1.txt`
  
- **Relative Path**: A path relative to the current working directory.
  - Example: `Documents/file1.txt`
  
#### **5. File Content Operations**
- **View File Contents**:
  - `cat <filename>`: Displays the contents of a file.
    - Example: `cat file1.txt`
  
- **Edit a File**:
  - `nano <filename>`: Opens a file for editing in the nano text editor.
    - Example: `nano file1.txt`
  
#### **6. System Information Commands**
- **Check Disk Space**:
  - `df -h`: Shows the disk usage of file systems in a human-readable format.
    - Example: `df -h`
  
- **Check Running Processes**:
  - `top`: Displays real-time system processes and resource usage.
  
- **System Uptime**:
  - `uptime`: Shows how long the system has been running.
#### **7. Grep Command**:
  - The grep command in Linux/Unix is used for searching specific patterns in files or output. It searches through files line by line and prints matching lines based on a regular expression pattern.
    - grep [options] pattern [file...]
    - grep "word" file.txt
    Searches for the string "word" in file.txt.
  
### **Directory Structure Explanation**
Explain the common directories in the Linux filesystem:

- **/**: The root directory, from where everything originates.
- **/home**: Contains personal directories for each user.
- **/bin**: Essential command binaries (like `ls`, `cp`, etc.).
- **/etc**: Configuration files for the system.
- **/var**: Variable data like logs and spools.
- **/usr**: User programs and system utilities.
